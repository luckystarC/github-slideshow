### **Problem 2. Matrix Operations**

**2.5 (1/4) Matrix Power**

Implement your own function first:

```python
# Problem 2.5 (a)
def mat_pow(A,p):
  # check input
  assert A.shape[0] == A.shape[1], "Matrix input needs to be square!"
  assert p > 0, "p should be a positive integer!"
  assert isinstance(p, int), "p should have the data type of integer!"

  ###### YOUR CODE STARTS HERE ######
  Ap = A
  for i in range(1,p):
    Ap=np.matmul(Ap,A) 
  ###################################
  return Ap
```

Call your own function to solve the given questions:

```python
print("Problem 2.5(b)")
###### YOUR CODE STARTS HERE ######

solution_5_b_1 = np.matmul(mat_pow(A,6),mat_pow(A,8)) # put (A^6)(A^8)here

solution_5_b_2 = mat_pow(A,6+8) # put A^(6+8) here 

relative_dist_5_b = LA.norm(solution_5_b_1-solution_5_b_2)/LA.norm(solution_5_b_1)

###################################
print(solution_5_b_1)
print()
print(solution_5_b_2)
print()
print("relative distance between these two matrices is: ")
print(relative_dist_5_b)
```

```python
print("Problem 2.5(c)")
###### YOUR CODE STARTS HERE ######

solution_5_c_1 =mat_pow(mat_pow(A,6),8)  # put (A^6)^8 here

solution_5_c_2 =mat_pow(A,6*8)  # put A^(6*8) here 

relative_dist_5_c = LA.norm(solution_5_c_1-solution_5_c_2)/LA.norm(solution_5_c_1)

###################################
print(solution_5_c_1)
print()
print(solution_5_c_2)
print()
print("relative distance between these two matrices is: ")
print(relative_dist_5_c)
print()
```

**2.6 (1/4) Inverse and Transpose**

```python
print("Problem 2.6(a)")
###### YOUR CODE STARTS HERE ######

solution_6_a_1 = LA.inv(np.matmul(A,D)) # put (AD)^(-1) here

solution_6_a_2 = np.matmul(LA.inv(D),LA.inv(A)) # put (D^(-1))(A^(-1)) here 

relative_dist_6_a = LA.norm(solution_6_a_1-solution_6_a_2)/LA.norm(solution_6_a_1)

###################################
print(solution_6_a_1)
print()
print(solution_6_a_2)
print()
print("relative distance between these two matrices is: ")
print(relative_dist_6_a)
print()
```

```python
print("Problem 2.6(b)")
###### YOUR CODE STARTS HERE ######
solution_6_b_1 = LA.inv(A).T # put (A^(-1))^T here

solution_6_b_2 = LA.inv(A.T) # put (A^T)^(-1) here 

relative_dist_6_b = LA.norm(solution_6_b_1-solution_6_b_2)/LA.norm(solution_6_b_1)

###################################
print(solution_6_b_1)
print()
print(solution_6_b_2)
print()
print("relative distance between these two matrices is: ")
print(relative_dist_6_b)
print()
```

```python
print("Problem 2.6(c)")
###### YOUR CODE STARTS HERE ######

solution_6_c_1 = np.matmul(A,B).T # put (AB)^(T) here (Updated)

solution_6_c_2 = np.matmul(B.T,A.T) # put (B^T)(A^T) here 

relative_dist_6_c = LA.norm(solution_6_c_1-solution_6_c_2)/LA.norm(solution_6_c_1)

###################################
print(solution_6_c_1)
print()
print(solution_6_c_2)
print()
print("relative distance between these two matrices is: ")
print(relative_dist_6_c)
```
